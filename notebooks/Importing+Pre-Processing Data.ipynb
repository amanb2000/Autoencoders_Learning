This module imports the celebrity faces dataset and saves it to a python called `"X_array"`.
## Dimension Details
* Images begin as 218*179 images.
* They are transformed into 45*35 images (first they are scaled down by a factor of 3 and then they are cropped in to only show faces). 
* The entire datastructure is a 202599 * (4725) datastructure (numpy array)

```python
def get_num_leading_zeros(big, small):
    size_big = 1
    size_small = 1
    
    if big == 0:
        big += 1;
    if small == 0:
        small += 1
    
    for i in range(12):
        if big / (10**i) < 1:
            size_big = i;
            break;
            
    for i in range(12):
        if small / (10**i) < 1:
            size_small = i;
            break;
    
    return size_big-size_small
```

```python
num_images = 202599
list_names = []
```

```python
for i in tqdm(range(num_images)):
    i = i+1
    num_zeros = get_num_leading_zeros(num_images, i)
    img_name = ''
    for j in range(num_zeros):
        img_name += '0'
    img_name += str(i)
    if(len(img_name) != 6):
        print("BAD: ",img_name)
    list_names.append(directory+img_name+'.jpg')
    
```

```python
# Function returns a numpy array version of an image of size (73, 59, 3)
def get_image(num):
    img_in = Image.open(list_names[num])
    im_resized = img_in.resize((59,73), Image.ANTIALIAS)
    test_image = np.asarray(im_resized)
    test_image = test_image[14:59, 12:47,:]
    return test_image
```

```python
# Testing conversion from picture to vector to picture

test_image = get_image(num_images-1)

test_image = test_image.flatten()
test_image = test_image.reshape(45,35,3)

plt.imshow(test_image)
```

```python
def show_img_vec(img_vec):
    img_vec_shaped = img_vec.reshape(45,35,3)
    plt.imshow(img_vec_shaped)
```

```python
X_narray = get_image(0).flatten()
one = np.array([1])

for i in tqdm(range(int(num_images))):
    img = get_image(i)
    img_flat = img.flatten()
    X_narray = np.vstack((X_narray, img_flat))
```
